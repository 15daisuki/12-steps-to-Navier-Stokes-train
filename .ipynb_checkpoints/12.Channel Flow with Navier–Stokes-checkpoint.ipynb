```python
# Step 12: Channel Flow with Navier–Stokes
# 第 12 步：纳维-斯托克斯方程下的通道流
# 向 u 动量方程中添加一个源项，以模拟压力驱动通道流的影响
	
import numpy
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D

# 定义压力方程右项 b
def build_up_b(rho, dt, dx, dy, u, v):
    b = numpy.zeros_like(u)
    b[1:-1, 1:-1] = (rho * (1 / dt * ((u[1:-1, 2:] - u[1:-1, 0:-2]) / (2 * dx) +
                                      (v[2:, 1:-1] - v[0:-2, 1:-1]) / (2 * dy)) -
                            ((u[1:-1, 2:] - u[1:-1, 0:-2]) / (2 * dx))**2 -
                            2 * ((u[2:, 1:-1] - u[0:-2, 1:-1]) / (2 * dy) *
                                 (v[1:-1, 2:] - v[1:-1, 0:-2]) / (2 * dx))-
                            ((v[2:, 1:-1] - v[0:-2, 1:-1]) / (2 * dy))**2))
    
    # 周期性假设 右边界的 “右边邻居” = 左边界的内部点（列索引为 0   
    # i,j b[1:-1, -1] 点处于右边界上x=2
    b[1:-1, -1] = (rho * (1 / dt * ((u[1:-1, 0] - u[1:-1, -2]) / (2 * dx) + 
                                    ((v[2:, -1] - v[0:-2, -1]) / (2 * dy)) - 
                            ((u[1:-1, 0] - u[1:-1, -2]) / (2 * dx))**2 - 
                             2 * ((u[2:, -1] - u[0:-2, -1]) / (2 * dy)) * 
                                  (v[1:-1, 0] - v[1:-1, -2]) / (2 * dx)) -
                             ((v[2:, -1] - v[0:-2, -1]) / (2 * dy))**2))

    # b[1:-1, 0] 点处于左边界上x=0
    b[1:-1, 0] = (rho * (1 / dt * ((u[1:-1, 1] - u[1:-1, -1]) / (2 * dx) + 
                                   (v[2:, 0] - v[0:-2, 0]) / (2 * dy)) - 
                         ((u[1:-1, 1] - u[1:-1, -1]) / (2 * dx))**2 - 
                         2 * ((u[2:, 0] - u[0:-2, 0]) / (2 * dy) * 
                              (v[1:-1, 1] - v[1:-1, -1]) / (2 * dx)) -
                         ((v[2:, 0] - v[0:-2, 0]) / (2 * dy))**2))

    return b

# 定义压力左项
def pressure_poisson_periodic(p, dx, dy):
    pn = numpy.empty_like(p)

    for q in range(nit):
        pn = p.copy()
        p[1:-1,1:-1] = (((pn[1:-1, 2:] + pn[1:-1, 0:-2]) * dy**2 + 
                        (pn[2:, 1:-1] + pn[0:-2, 1:-1]) * dx**2) /
                        (2 * (dx**2 + dy**2)) - 
                        dx**2 * dy**2 / (2 * (dx**2 + dy**2)) * b[1:-1, 1:-1])

        # x = 2
        p[1:-1, -1] = (((pn[1:-1, 0] + pn[1:-1, -2]) * dy**2 + 
                        (pn[2:, -1] + pn[0:-2, -1]) * dx**2) /
                        (2 * (dx**2 + dy**2)) - 
                        dx**2 * dy**2 / (2 * (dx**2 + dy**2)) * b[1:-1, -1])

        # x = 0
        p[1:-1, 0] = (((pn[1:-1, 1] + pn[1:-1, -1]) * dy**2 + 
                        (pn[2:, 0] + pn[0:-2, 0]) * dx**2) /
                        (2 * (dx**2 + dy**2)) - 
                        dx**2 * dy**2 / (2 * (dx**2 + dy**2)) * b[1:-1, 0])
        
        # 边界层
        p[-1, :] = p[-2, :]  # dp/dy = 0 at y = 2
        p[0, :] = p[1, :]    # dp/dy = 0 at y = 0

    return p
        

nx = 41
ny = 41
c = 1
nt = 10
nit = 50
dx = 2 / (nx -1)
dy = 2 / (ny -1)
x = numpy.linspace(0, 2, nx)
y = numpy.linspace(0, 2, ny)
X, Y = numpy.meshgrid(x, y)

rho = 1
nu = 0.1
F = 1
dt = 0.01

u = numpy.zeros((ny,nx))
un = numpy.zeros((ny,nx))

v = numpy.zeros((ny,nx))
vn = numpy.zeros((ny,nx))

p = numpy.zeros((ny,nx))
pn = numpy.zeros((ny,nx))

b = numpy.zeros((ny,nx))

udiff = 1
stepcount = 0

while udiff > 0.001:
    un = u.copy()
    vn = v.copy()

    b = build_up_b(rho, dt, dx, dy, u, v)
    p = pressure_poisson_periodic(p, dx, dy)

    u[1:-1, 1:-1] = (un[1:-1, 1:-1] -
                     un[1:-1, 1:-1] * dt / dx * 
                    (un[1:-1, 1:-1] - un[1:-1, 0:-2]) -
                     vn[1:-1, 1:-1] * dt / dy * 
                    (un[1:-1, 1:-1] - un[0:-2, 1:-1]) -
                     dt / (2 * rho * dx) * 
                    (p[1:-1, 2:] - p[1:-1, 0:-2]) +
                     nu * (dt / dx**2 * 
                    (un[1:-1, 2:] - 2 * un[1:-1, 1:-1] + un[1:-1, 0:-2]) +
                     dt / dy**2 * 
                    (un[2:, 1:-1] - 2 * un[1:-1, 1:-1] + un[0:-2, 1:-1])) + 
                     F * dt)
    # x = 2
    u[1:-1, -1] = (un[1:-1, -1] -
                     un[1:-1, -1] * dt / dx * 
                    (un[1:-1, -1] - un[1:-1, -2]) -
                     vn[1:-1, -1] * dt / dy * 
                    (un[1:-1, -1] - un[0:-2, -1]) -
                     dt / (2 * rho * dx) * 
                    (p[1:-1, 0] - p[1:-1, -2]) +
                     nu * (dt / dx**2 * 
                    (un[1:-1, 0] - 2 * un[1:-1, -1] + un[1:-1, -2]) +
                     dt / dy**2 * 
                    (un[2:, -1] - 2 * un[1:-1, -1] + un[0:-2, -1])) + 
                     F * dt)
    # x = 0
    u[1:-1, 0] = (un[1:-1, 0] -
                     un[1:-1, 0] * dt / dx * 
                    (un[1:-1, 0] - un[1:-1, -1]) -
                     vn[1:-1, 0] * dt / dy * 
                    (un[1:-1, 0] - un[0:-2, 0]) -
                     dt / (2 * rho * dx) * 
                    (p[1:-1, 1] - p[1:-1, -1]) +
                     nu * (dt / dx**2 * 
                    (un[1:-1, 1] - 2 * un[1:-1, 0] + un[1:-1, -1]) +
                     dt / dy**2 * 
                    (un[2:, 0] - 2 * un[1:-1, 0] + un[0:-2, 0])) + 
                     F * dt)

    v[1:-1, 1:-1] = (vn[1:-1, 1:-1] -
                     un[1:-1, 1:-1] * dt / dx * 
                    (vn[1:-1, 1:-1] - vn[1:-1, 0:-2]) -
                     vn[1:-1, 1:-1] * dt / dy * 
                    (vn[1:-1, 1:-1] - vn[0:-2, 1:-1]) -
                     dt / (2 * rho * dy) * 
                    (p[2:, 1:-1] - p[0:-2, 1:-1]) +
                     nu * (dt / dx**2 *
                    (vn[1:-1, 2:] - 2 * vn[1:-1, 1:-1] + vn[1:-1, 0:-2]) +
                     dt / dy**2 * 
                    (vn[2:, 1:-1] - 2 * vn[1:-1, 1:-1] + vn[0:-2, 1:-1])))

    # x = 2 
    v[1:-1, -1] = (vn[1:-1, -1] - un[1:-1, -1] * dt / dx *
                  (vn[1:-1, -1] - vn[1:-1, -2]) - 
                   vn[1:-1, -1] * dt / dy *
                  (vn[1:-1, -1] - vn[0:-2, -1]) -
                   dt / (2 * rho * dy) * 
                  (p[2:, -1] - p[0:-2, -1]) +
                   nu * (dt / dx**2 *
                  (vn[1:-1, 0] - 2 * vn[1:-1, -1] + vn[1:-1, -2]) +
                   dt / dy**2 *
                  (vn[2:, -1] - 2 * vn[1:-1, -1] + vn[0:-2, -1])))

    # x = 0
    v[1:-1, 0] = (vn[1:-1, 0] - un[1:-1, 0] * dt / dx *
                 (vn[1:-1, 0] - vn[1:-1, -1]) -
                  vn[1:-1, 0] * dt / dy *
                 (vn[1:-1, 0] - vn[0:-2, 0]) -
                  dt / (2 * rho * dy) * 
                 (p[2:, 0] - p[0:-2, 0]) +
                  nu * (dt / dx**2 * 
                 (vn[1:-1, 1] - 2 * vn[1:-1, 0] + vn[1:-1, -1]) +
                  dt / dy**2 * 
                 (vn[2:, 0] - 2 * vn[1:-1, 0] + vn[0:-2, 0])))
    # 设置速度 u,v = 0 at y = 0,2
    u[0, :] = 0
    u[-1, :] = 0
    v[0, :] = 0
    v[-1, :] = 0

    udiff = (numpy.sum(u) - numpy.sum(un)) / numpy.sum(u)
    stepcount += 1 # 迭代次数
    
print(stepcount)

# 二维压力驱动通道流（泊肃叶流）的速度场可视化图
fig = pyplot.figure(figsize = (11,7), dpi=100)
pyplot.quiver(X[::3, ::3], Y[::3, ::3], u[::3, ::3], v[::3, ::3]);

contf = pyplot.contourf(X, Y, u, alpha=0.5, cmap='viridis')
cbar = pyplot.colorbar(contf)
cbar.set_label('X direction u velocity', rotation=0, fontsize=12, labelpad=10, ha='center') 
pyplot.xlabel('X', fontsize=12)
pyplot.ylabel('Y', rotation=0, fontsize=12, labelpad=25, ha='center')
pyplot.title('Channel Flow with Navier–Stokes', fontsize=14)
pyplot.show()

```
