```python
import numpy
from matplotlib import pyplot

# Step 3: Diffusion Equation in 1-D 一维扩散方程
# 空间域 0~2
# 时间域 0~nt*dt
nx = 41
dx = 2 / (nx -1)   # 空间步长
nt = 20            # 时间步进次数
nu = 0.3  # viscosity 扩散系数
sigma = 0.2
dt = sigma * dx**2 / nu # 时间步长（由稳定性条件决定）由 CFL<sigma

u = numpy.ones(nx)
u[int(0.5 / dx):int(1 /dx + 1)] = 2

un = numpy.ones(nx)

for n in range(nt):
    un = u.copy()
    for i in range(1, nx-1):
        u[i] = un[i] + nu * dt / dx**2 * (un[i+1] - 2 * un[i] + un[i-1])

pyplot.plot(numpy.linspace(0, 2, nx), u);

        
```
